```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import utils_kernel as KT 
import HS_utils as utils
import data_generator as generator
```

```python
torch.set_default_dtype(torch.float64)
```

```python
mu_ex= generator.mu_ex
sigma_ex = generator.sigma_ex
#sigma_ex = generator.sigma_aux(3)

n = 200
(x,y) = generator.generateData(mu_ex,sigma_ex,n)
```

```python
kern_e = KT.gaussianKernel(0.1)
la_e = 0.05
LME = utils.LMK(kern_e,x,la_e)
```

```python
kern_l = KT.gaussianKernel(0.2)
la_l = 0.1
LML = utils.QKM(kern_l,x,la_l,positive = True)
#LML = utils.LMK(kern_l,x,la_l,positive = 'coefficients')
```

```python
lmodel = utils.HSLinearModel(LME,LML)
```

```python
loss = utils.LossHS(y)
```

```python
model = utils.HSModel(loss,lmodel)
```

```python
Niter = 1000
freq = Niter//10
cb,cobj = model.cbcboj_pd(freq,plot= True)

res = model.prox_method(Niter,cb = cb,cobj = cobj)
```

```python
xtest = torch.linspace(0,1,300)
rtest = model.Rx_dual(res,xtest)
eta = rtest[0]
la = rtest[1]
plt.plot(xtest,eta)
plt.show()
plt.plot(xtest,la)
plt.show()
plt.plot(xtest,eta/la,'b')
plt.plot(xtest,eta/la+1.96*torch.sqrt(1/la),color = 'orange')
plt.plot(xtest,eta/la-1.96*torch.sqrt(1/la),color = 'orange')
plt.plot(xtest,mu_ex(xtest),'r')
plt.plot(x,mu_ex(x)+1.96*sigma_ex(x),'g',linestyle = '--')
plt.plot(x,mu_ex(x)-1.96*sigma_ex(x),'g',linestyle ='--')
plt.show()
```
