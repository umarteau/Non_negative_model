```python
# Number of data points we will be using and basic parameters of the model
ndata = 80
# data generator : toy example of smola et al
mu_ex= generator.mu_ex
sigma_ex = generator.sigma_ex

eta_ex = lambda x : mu_ex(x)/sigma_ex(x)**2
la_ex = lambda x : 1/sigma_ex(x)**2

def gen(n = ndata):
    mu_ex= generator.mu_ex
    sigma_ex = generator.sigma_ex
    #sigma_ex = generator.sigma_aux(3)
    (x,y) = generator.generateData(mu_ex,sigma_ex,n,plot = False)
    return (x,y)

data_set = gen()
xdata_path = pt_path + 'xdataExperiments.pt'
ydata_path = pt_path + 'ydataExperiments.pt'

torch.save(data_set[0].view(ndata,1),xdata_path)
torch.save(data_set[1].view(ndata),ydata_path)


def data_split(data_set,r=0.5):
    x,y = data_set
    n = x.size(0)
    l = np.array(range(n))
    np.random.shuffle(l)
    x,y = x[l,:],y[l]
    ntr = int(np.ceil(n*r))
    nts = n - ntr
    data_set_train = x[:ntr,:],y[:ntr]
    data_set_test =x[ntr:,:],y[ntr:]
    return data_set_train,data_set_test



def Loss0(rx,ntest = ndata,data_set_test = None):
    if isinstance()
    
    #Here eps is the moreau smoothing parameter
    if isinstance(data_set_test ,type(None)):
        xtest,ytest = gen(n = ntest)
    else:
        xtest,ytest = data_set_test
        ntest = xtest.size(0)
    
    r = rx(xtest)
    if eps == 0:
        if (r[1]<=0).sum() > 0:
            return torch.tensor(np.inf)
        else:
            LL = utils.HSlossNW(ytest,eps).L
            L = lambda al : LL(al)/ntest
            rxe =  rx(xtest)
            rxe = [rxe[0].view(ntest,1),rxe[1].view(ntest,1)]
            return L(rxe)
            
    else:
        Leps = utils.HSlossNW(ytest,eps).Leps
        L = lambda al : Leps(al)/ntest
        rxe =  rx(xtest)
        rxe = [rxe[0].view(ntest,1),rxe[1].view(ntest,1)]
        return L(rxe)

```

# Experiments using basic linear model

```python
def makerx(data_set,lae,lal,sigmae = 0.3,sigmal = 0.4,c=0,Niter = 100):
    x,y = data_set
    my = y.mean()
    yr = y-my
    
    
    LME = utils.LMK(sigmae,x,lae,kernel = 'gaussian',centered = False,c = c)
    LML = utils.LMK(sigmal,x,lal,kernel = 'gaussian',centered = False,c = c)
    
    lmodel = utils.HSLinearModel(LME,LML)
    model = utils.HSModel(lmodel,yr)
    
    freq = Niter//5
    #cb,cobj = model.cbcboj_pd(freq,plot= True)

    #al = model.prox_method(Niter,cb = cb,cobj = cobj)
    al = model.prox_method(Niter)
    
    def rx(x):
        rr = model.Rx_dual(al,x)
        eta = rr[0]
        la = rr[1]
        return [eta + my*la,la]
    return rx
```

# Our model 

```python
def makerx(data_set,lae,lal,sigmae = 0.3,sigmal = 0.4,c=0,Niter = 1000,show_plot = False):
    x,y = data_set
    my = y.mean()
    yr = y-my
    

    LME = utils.LMK(sigmae,x,lae,kernel = 'gaussian',centered = False,c = c)
    
    err = 0.01
    LML = utils.QKM(sigmal,x,lal*err,mu = lal,kernel = 'gaussian',centered = False,c = 0,positive = True)
    
    lmodel = utils.HSLinearModel(LME,LML)
    model = utils.HSModel(lmodel,yr)
    
    if show_plot:
        freq = Niter//5
        cb,cobj = model.cbcboj_pd(freq,plot= True)
        al = model.prox_method(Niter,cb = cb,cobj = cobj)
    else:
        al = model.prox_method(Niter)
    
    def rx(x):
        rr = model.Rx_dual(al,x)
        eta = rr[0]
        la = rr[1]
        return [eta + my*la,la]
    return rx
```

## Finding the right parameters 

```python
lae_l = [0.01,0.001]
lal_l = [0.01,0.001]
sigmae_l = [0.3]
sigmal_l = [0.4]

param_l = []
for lae in lae_l:
    for lal in lal_l:
        if lae/lal < 10.1 and lal/lae < 10.1:
            for sigmae in sigmae_l:
                for sigmal in sigmal_l:
                    param_l.append((lae,lal,sigmae,sigmal))
        

NN = len(param_l)

res = torch.zeros((5,NN))
res[1:,:] = torch.tensor(param_l).T

r = 0.7
K = 5
res_list = []
data_set = torch.load(xdata_path),torch.load(ydata_path)
for i in range(NN):
    for k in range(K):
        data_set_train,data_set_test = data_split(data_set,r=r)
        lae,lal,sigmae,sigmal = res[1,i],res[2,i],res[3,i],res[4,i]
        rx = makerx(data_set,lae,lal,sigmae = sigmae,sigmal = sigmal,Niter = 50 + 25*int(np.sqrt(1/min(lae,lal))))
        print('lambda_e = {}, lambda_l = {}, sigmae = {},sigmal = {}'.format(lae,lal,sigmae,sigmal))
        ll = LossEps(rx)
        #ll = squareloss(px,data_set_test = data_set_test,ntest = 0,ntestbis = 3000)
        print("ll = {}".format(ll))
        res[0,i] += ll
    res[0,i]/= K
```

# Nadaraya Watson stuff

```python
def makerx(data_set,lae,lal,sigmae = 0.1,sigmal = 0.1,c=0,Niter = 50,show_plot = False):
    x,y = data_set
    my = y.mean()
    yr = y-my
    
    eps = 0.01
    loss = utils.HSlossNW(yr,eps)
    
    
    LME = utils.kernelModel(sigmae,x,kernel = 'gaussian')
    LML = utils.kernelModel(sigmal,x,kernel = 'gaussian',positive = True,c=c)
    model = utils.HSLinearModelNW(LME,LML)
    
    hsm = utils.HSModelNW(loss,lae,lal,model)
    
    al = hsm.FISTA(Niter)
    
    def rx(x):
        rr = model.Rx(al,x)
        eta = rr[0]
        la = rr[1]
        return [eta + my*la,la]
    return rx
```

# Finding the right parameters

```python
lae_l = [0.1,0.01,0.001]
lal_l = [0.1,0.01,0.001]
la_l = []
for lae in lae_l:
    for lal in lal_l:
        la_l.append((lae,lal))
        

K = 2
res_list = []
for la in la_l:
    lae = la[0]
    lal = la[1]
    res = 0
    for k in range(K):
        data_set = gen()
        rx = makerx(data_set,lae,lal,Niter = 100*int(np.sqrt(1/min(lae,lal))))
        print('lambda_e = {}, lambda_l = {}'.format(lae,lal))
        res += LossEps(rx)
    res_list.append(res/K)
```

```python
res_l = torch.zeros((3,len(la_l))) 
res_l[2,:] = torch.tensor(res_list)
res_l[0,:] = torch.tensor([v[0] for v in la_l])
res_l[1,:] = torch.tensor([v[1] for v in la_l])
torch.save(res_l,'lambda_choice_NW0.pt')
```

# GLM

```python
def makerx(data_set,lae,lal,sigmae = 0.3,sigmal = 0.4,c=0,Niter = 50,show_plot = False):
    x,y = data_set
    my = y.mean()
    yr = y-my
    
    eps = 0.01
    loss = utils.HSlossNW(yr,eps)
    
    
    LME = utils.kernelExpoModel(sigmae,x,kernel = 'gaussian',c=c,positive = False)
    LML = utils.kernelExpoModel(sigmal,x,kernel = 'gaussian',positive = True)
    model = utils.HSLinearModelExpo(LME,LML)
    
    hsm = utils.HSModelExpo(loss,lae,lal,model)
    
    al = hsm.GD(Niter)
    
    def rx(x):
        rr = model.px(al,x)
        eta = rr[0]
        la = rr[1]
        return [eta + my*la,la]
    return rx
```

FInding $\lambda$

```python
lae_l = [1,0.1,0.01,0.001,0.0001,0.00001]
lal_l = [1,0.1,0.01,0.001,0.0001,0.00001]
la_l = []
for lae in lae_l:
    for lal in lal_l:
        la_l.append((lae,lal))
        

K = 10
res_list = []
for la in la_l:
    lae = la[0]
    lal = la[1]
    res = torch.tensor(np.inf)
    for k in range(K):
        data_set = gen()
        Niter = 25+50*int(np.sqrt(1/min(lae,lal)))
        rx = makerx(data_set,lae,lal,Niter = 1)
        print('lambda_e = {}, lambda_l = {}'.format(lae,lal))
        print('lossEps {}'.format(LossEps(rx)))
        res = min(LossEps(rx),res)
    res_list.append(res)
```

```python
res_l = torch.zeros((3,len(la_l))) 
res_l[2,:] = torch.tensor(res_list)
res_l[0,:] = torch.tensor([v[0] for v in la_l])
res_l[1,:] = torch.tensor([v[1] for v in la_l])
torch.save(res_l,'lambda_choice_GLM0.pt')
```
